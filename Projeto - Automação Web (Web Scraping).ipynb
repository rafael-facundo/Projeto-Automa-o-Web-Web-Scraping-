# Automação Web e Busca de Informações com Python

#### Desafio: 

Trabalhamos em uma importadora e o preço dos nossos produtos é vinculado a cotação de:
- Dólar
- Euro
- Ouro

Precisamos pegar na internet, de forma automática, a cotação desses 3 itens e saber quanto devemos cobrar pelos nossos produtos, considerando uma margem de contribuição que temos na nossa base de dados.

Base de Dados: https://drive.google.com/drive/folders/1KmAdo593nD8J9QBaZxPOG1yxHZua4Rtv?usp=sharing

Para isso, vamos criar uma automação web:

- Usaremos o selenium
- Importante: baixar o webdriver

```python
!pip install selenium
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

# passo 1: pegar cotação dólar
# abrir o navegador
nav=webdriver.Chrome()

# entrar no google
nav.get('https://www.google.com/')

# pesquisar cotação
nav.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('cotação dolar')
nav.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)

# pegar a cotação
cot_dol=nav.find_element('xpath','//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cot_dol)

# passo 2: pegar cotação euro
nav.get('https://www.google.com/')
nav.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys('cotação euro')
nav.find_element('xpath','/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input').send_keys(Keys.ENTER)
cot_euro=nav.find_element('xpath','//*[@id="knowledge-currency__updatable-data-column"]/div[1]/div[2]/span[1]').get_attribute('data-value')
print(cot_euro)


# passo 3: pegar cotação ouro
nav.get('https://www.melhorcambio.com/ouro-hoje#:~:text=O%20valor%20do%20grama%20do,em%20R%24%20293%2C35.')
cot_ouro=nav.find_element('xpath','//*[@id="comercial"]').get_attribute('value')
cot_ouro=cot_ouro.replace(',', '.')
print(cot_ouro)

nav.quit()




```

### Agora vamos atualiza a nossa base de preços com as novas cotações

- Importando a base de dados

```python
# passo 4: atualizar a base de dados
import pandas as pd
tab=pd.read_excel('Produtos.xlsx')
display(tab)
```

- Atualizando os preços e o cálculo do Preço Final

```python
# passo 5: recalcular os preços
# atualizar as cotações
tab.loc[tab['Moeda']=='Dólar', 'Cotação'] = float(cot_dol)
tab.loc[tab['Moeda']=='Euro', 'Cotação'] = float(cot_euro)
tab.loc[tab['Moeda']=='Ouro', 'Cotação'] = float(cot_ouro)

# preço de compra = preço original * cotação
tab['Preço de Compra'] = tab['Preço Original'] * tab['Cotação']

# preço de venda = preço de compra * margem
tab['Preço de Venda'] = tab['Preço de Compra'] * tab['Margem']

display(tab)
```

### Agora vamos exportar a nova base de preços atualizada

```python
# passo 6: exportar a base de dados
tab.to_excel('Produtos Novo.xlsx', index=False)
```
